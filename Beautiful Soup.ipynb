```python
# Some Imports. Did I mention that the `requests` library is great?
import requests

# This stands for `beautifulsoup4`. Also great.
import bs4
```

```python
# We're going to scrape some Wikipedia pages. First
# we need URLs. Let's find out what Wikipedia has to
# say about Babylon 5. 

all_babylon = 'https://en.wikipedia.org/wiki/Index_of_Babylon_5_articles'

# Fetch the page using `requests.get` and get the content
# using the `text` attribute:

html_text = requests.get(all_babylon).text

# Pass the result to bs4, which will parse the HTML.

html = bs4.BeautifulSoup(html_text, 'html.parser')

# To get the htmls we need to fetch the `a` tags. 
# We can do that by passing a list of tag names to 
# `html` as if it were a function.

urls = [link.get('href', '') for link in html(['a'])]
print('We have downloaded urls for {} pages'.format(len(urls)))
```

```python
# For now we'll just pick a single URL...

spoo = [u for u in urls if "Spoo" in u]  # Don't ask.
spoo_url = 'https://en.wikipedia.org' + spoo[0]

# Again, we pass the url to `requests.get`...

spoo_html_text = requests.get(spoo_url).text

# And have bs4 parse it.

spoo_html = bs4.BeautifulSoup(spoo_html_text, 'html.parser')

# But this time we want to make sure we're only
# getting text. We don't want javascript or css
# mucking up our text, so we go through the
# script and style tags and `extract` them.
for tag in spoo_html(['script', 'style']):
    discarded_tag = tag.extract()

print(spoo_html.get_text())
```
